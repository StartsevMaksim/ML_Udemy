```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = pd.read_csv('dm_office_sales.csv')
```

```python
#sns.scatterplot(x=data_1['sales'], y=data_1['salary']);
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(data=data_1, x='salary', y='sales', s=200, style='level of education', hue='level of education');
plt.savefig('test_fig_1.jpeg')
```

```python
sns.rugplot(data_1, x='salary', height=0.5)
```

```python
sns.set(style='dark')
sns.displot(data_1, x='salary', bins=30, color='red', linewidth=4, edgecolor = 'blue', kde=True);
```

```python
sns.kdeplot(data=data_1, x='salary', bw_adjust=0.5, fill=True);
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data_1, x='division', hue='division', palette='Dark2');
```

```python
plt.figure(figsize=(10,6), dpi=200)
sns.barplot(data=data_1, x='level of education', y='salary', estimator=np.mean, errorbar='sd', hue='division')
plt.legend(loc=(1.07,0.7));
```

```python
data_2 = pd.read_csv('StudentsPerformance.csv')
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=data_2, y='math score', x='gender', hue='test preparation course', palette='Dark2')
plt.legend(loc=(1.02, 0.5));
```

```python
sns.violinplot(data=data_2, y='math score', x='gender', 
               hue='test preparation course', palette='Dark2', inner='quartile', bw_adjust=0.6)
plt.legend(loc=(1.02, 0.5))
```

```python
sns.swarmplot(data=data_2, x='math score', y='gender', size=2);
```

```python
sns.jointplot(data=data_2,x='math score', y='reading score', kind='hex');
```

```python
data_2_pair = data_2[['math score', 'writing score']]
sns.pairplot(data_2_pair);
```
