```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg #.jpg, .png -> Массив numpy
from sklearn.cluster import KMeans
```

```python
image_as_array = mpimg.imread('../DATA/palm_trees.jpg')
```

```python
h, w, c = image_as_array.shape #H, W, C
```

```python
plt.figure(dpi=200)
plt.imshow(image_as_array)
```

```python
colors = np.array([color for width in image_as_array for color in width])
      #= image_as_array.reshape(H*W, C)
```

```python
colors.shape
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(colors) #Номера кластеров для каждого пикселя
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int) #Центры(цвета) кластеров
```

```python
#Сопоставление цветов кластеров к их номерам для каждого пикселя
rgb_codes = rgb_codes[labels]
```

```python
quantized_image = np.array([[rgb_codes[index] for index in range(row*w, (row+1)*w)] for row in range(h)])
```

```python
plt.figure(dpi=200)
plt.imshow(quantized_image)
```
