```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
df = pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_dummy = pd.get_dummies(df.drop('name', axis=1))
```

```python
df_dummy
```

```python
scaler = MinMaxScaler() #Для иерархической кластеризации
```

```python
scaled_X = scaler.fit_transform(df_dummy)
```

```python
scaled_df = pd.DataFrame(data=scaled_X, columns=df_dummy.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, row_cluster=False)
#Высота линии отвечает за расстояние между кластерами
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, col_cluster=False)
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
df.columns
```

```python
plt.figure(figsize=(14,6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='horsepower', hue=cluster_labels, palette='viridis')
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=2) #при расстоянии между точками >= 
#distance_threshold точки не будут объединяться в кластеры
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
np.set_printoptions(suppress=True)
```

```python
linkage_matrix = hierarchy.linkage(scaled_df.values, method='ward')
```

```python
linkage_matrix #[[Кластер_1, Кластер_2, Расстояние, Кол-во точек в кластерах]]
```

```python
plt.figure(figsize=(20,10), dpi=200)
dend = dendrogram(linkage_matrix, truncate_mode='lastp', p=11)
plt.xticks(rotation=90);
```
