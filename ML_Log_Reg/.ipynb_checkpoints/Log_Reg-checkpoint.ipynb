```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay, \
                            PrecisionRecallDisplay, RocCurveDisplay
```

```python
data_1 = pd.read_csv('../DATA/hearing_test.csv')
```

```python
data_1.head()
```

```python
sns.scatterplot(data_1, x='physical_score', y='age', hue='test_result', alpha=0.7);
```

```python
sns.heatmap(data_1.corr(), annot=True);
```

```python
X = data_1.drop('test_result', axis=1)
y = data_1['test_result']
```

```python
#   -
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
#    (ROC-)
```

```python
accuracy_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
print(classification_report(y_test, y_predict)) #precision -        
                                                #       
                                                #recall -         
                                                #   
```

```python
RocCurveDisplay.from_estimator(model, X_test, y_test)
```

```python
PrecisionRecallDisplay.from_estimator(model, X_test, y_test)
```

```python
iris_data = pd.read_csv('../DATA/iris.csv')
```

```python
iris_data
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris_data, hue='species') # setosa    
```

```python
sns.pairplot(iris_data, hue='species')
```

```python
X = iris_data.drop('species', axis=1)
y = iris_data['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(solver='saga',     # 
                           multi_class='ovr', #multi_class -     .
                           max_iter=5000)     #ovr -       ,     
```

```python
penalty = ['l1', 'l2', 'elasticnet']
l1_ratio = np.linspace(0, 1, 20)
C = np.logspace(0, 10, 20)

parama_grid = {'penalty': penalty,
               'l1_ratio': l1_ratio,
               'C': C}
```

```python
grid_model = GridSearchCV(model, param_grid=parama_grid)
```

```python
grid_model.fit(X_train, y_train)
```

```python
y_predict = grid_model.predict(X_test)
```

```python
accuracy_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```

```python
ConfusionMatrixDisplay.from_estimator(grid_model, X_test, y_test)
```

```python
print(classification_report(y_test, y_predict))
```
