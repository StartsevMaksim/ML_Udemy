```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_arr_1 = np.arange(0, 10, 0.1)
y_arr_1 = x_arr_1 ** 3
x_arr_2 = np.linspace(0, 10, 30)
y_arr_2 = x_arr_2 * 4
```

```python
plt.plot(x_arr_1, y_arr_1)
plt.ylabel('Ось Oy')
plt.xlabel('Ось Ox')
plt.title('Name')
plt.savefig('Figure_1.jpeg')
```

```python
fig = plt.figure(figsize = (6,4),dpi=100)
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(x_arr_1, y_arr_1)
axes_1.set_xlabel('Ось Ox')
axes_1.set_ylabel('Ось Oy')
axes_1.set_title('Рост');
axes_2 = fig.add_axes([0.1, 0.5, 0.3, 0.3])
axes_2.plot(x_arr_2, y_arr_2)
axes_2.set_xlabel('Ось Ox')
axes_2.set_ylabel('Ось Oy')
axes_2.set_title('Производная');
```

```python
fig.savefig('Exp_fig.jpeg', bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x_arr_2, y_arr_2)
axes[1].plot(x_arr_1, y_arr_1)
fig.subplots_adjust(wspace=0.9, hspace=0.5)
```

```python
fig = plt.figure(figsize = (6,4),dpi=100)
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(x_arr_1, y_arr_1, label='Х в квадрате')
axes_1.plot(x_arr_2, y_arr_2, label='Линейная')
axes_1.set_xlabel('Ось Ox')
axes_1.set_ylabel('Ось Oy')
axes_1.set_title('Рост')
axes_1.legend(loc=(1.1, 1));
```

```python
fig = plt.figure(figsize = (6,4),dpi=100)
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(x_arr_1, y_arr_1, color='red', lw=0.5, ls='--', marker = '|')
axes_1.set_xlabel('Ось Ox')
axes_1.set_ylabel('Ось Oy')
axes_1.set_title('Рост');
```
