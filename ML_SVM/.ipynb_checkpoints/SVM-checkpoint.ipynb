```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC, SVR, LinearSVR
from svm_margin_plot import plot_svm_boundary
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
help(SVC) # C - часть точек, которые могут быть классифицированы неверно 
          # kernel - тип ядра
```

```python
df = pd.read_csv('../DATA/mouse_viral_study.csv')
```

```python
df.head()
```

```python
X = df.drop('Virus Present', axis=1)
y = df['Virus Present']
```

```python
model = SVC(kernel='linear', C=1000) # C = 1000 - значит малое число ошибок
```

```python
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=0.05)
```

```python
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='rbf', C=1, gamma='auto') #gamma - влияние каждой точки на результат
```

```python
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='sigmoid')
```

```python
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='poly', degree=2) #degree - степень полинома
```

```python
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
svm = SVC()
```

```python
param_grid = {'C': [0.01, 0.1, 1], 'kernel': ['linear', 'rbf']}
```

```python
grid = GridSearchCV(svm, param_grid=param_grid)
```

```python
grid.fit(X, y)
```

```python
grid.best_params_
```

```python
#SVM для регрессии
#Мы строим нашу гиперплоскость, а потом по значениям признаков определяем точку результат по нашей гиперплоскости
```

```python
df = pd.read_csv('../DATA/cement_slump.csv')
```

```python
df.head()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
X = df.drop('Compressive Strength (28-day)(Mpa)', axis=1)
y = df['Compressive Strength (28-day)(Mpa)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = SVR()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
np.sqrt(mean_squared_error(y_test, y_predict))
```

```python
param_grid = {'C': [0.001, 0.01, 0.1, 0.5, 1],
              'kernel': ['linear', 'rbf', 'poly'],
              'gamma': ['scale', 'auto'],
              'degree': [2, 3, 4],
              'epsilon': [0, 0.01, 0.1, 0.5, 1, 2]}
```

```python
model = SVR()
```

```python
grid = GridSearchCV(model, param_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_predict = grid.predict(X_test)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
np.sqrt(mean_squared_error(y_test, y_predict))
```
