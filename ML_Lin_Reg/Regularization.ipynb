```python
#Применение разных типов регуляризации
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from UserAlgos import UStandardScaler
from sklearn.linear_model import Ridge, RidgeCV, LassoCV, ElasticNetCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
data_1 = pd.read_csv('Advertising.csv')
```

```python
X = data_1.drop('sales', axis=1)
y = data_1['sales']
```

```python
poly_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_features = poly_converter.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=42)
```

```python
#Выполняем масштабирование признаков
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
#Отмасштабированые признаки
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
#L2-регуляризация (без кросс-валидации)
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
RMSE
```

```python
#L2-регуляризация (с кросс-валидацией)
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0), scoring='neg_mean_absolute_error') #По какой метрике
                                                                                     #будут выбираться лучшие значения Alpha
```

```python
ridge_cv_model.fit(X_train, y_train)
```

```python
ridge_cv_model.alpha_ #Лучший вариант Alpha
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
RMSE
```

```python
#L1-регуляризация (с кросс-валидацией)
```

```python
lasso_cv_model = LassoCV(eps=0.001, n_alphas=100, cv=5, max_iter=1000000)
```

```python
lasso_cv_model.fit(X_train, y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
RMSE
```

```python
#L3-регуляризация (с кросс-валидацией)
```

```python
elastic_cv_model = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1],
                               eps=0.001, n_alphas=100, max_iter=1000000)
```

```python
elastic_cv_model.fit(X_train, y_train)
```

```python
elastic_cv_model.l1_ratio_
```

```python
test_predictions = elastic_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
RMSE
```
