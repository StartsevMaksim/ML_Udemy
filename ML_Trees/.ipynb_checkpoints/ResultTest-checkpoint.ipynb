```python
data_dummies = data.drop('customerID', axis=1)
```

```python
data_dummies = pd.get_dummies(data_dummies, drop_first=True)
```

```python
X = data_dummies.drop('Churn_Yes', axis=1)
y = data_dummies['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
model = RandomForestClassifier()
```

```python
param_grid = {'n_estimators': [10,50,100,200,300],
              'max_features': [2,5,7,8]}
```

```python
grid = GridSearchCV(model, param_grid, verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
y_predict = grid.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
```

```python
ConfusionMatrixDisplay.from_estimator(grid.best_estimator_, X_test, y_test)
```

```python
feature_imp = pd.DataFrame(data=grid.best_estimator_.feature_importances_,
                           index=grid.best_estimator_.feature_names_in_,
                           columns=['Важность']).sort_values('Важность')
```

```python
plt.figure(figsize=(14,6), dpi=200)
sns.barplot(data=feature_imp, x=feature_imp.index, 
            y='Важность', 
            hue=feature_imp.index)
plt.xticks(rotation=90);
```
